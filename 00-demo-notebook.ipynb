## Pushing to GitHub from this environment

This environment is **ephemeral**.  
To push code or notebooks to GitHub, authenticate **inside the notebook**.

### Recommended: GitHub CLI (`gh`)

```bash

```

2. Login:

3. Clone and work inside your repo:
